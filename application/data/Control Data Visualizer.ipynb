```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("exp1_3-8-22/controltarg_7_29.csv")
df = pd.read_csv("exp1_3-8-22/controltarg_7_29_from-7_29.csv")
df
```

```python
marker_set = ["M1", "M2", "M3", "M4", "M5", "M6", "M7", "M8", "M9", "M10"]

target = (7,29)
finalx = round(df["M10X"].iloc[-1], 3)
finaly = round(df["M10Y"].iloc[-1], 3)
final = (finalx, finaly)

plt.plot([0],[0], "o", label="(0,0)")
plt.plot(finalx, finaly, "x", label="Targ", color="red")
for m in marker_set:
    x_key = m + "X"
    y_key = m + "Y"
    x = df[x_key]
    y = df[y_key]
    plt.plot(x,y, label=m)
y_max = 35
x_max = y_max / 2
plt.title("End-Effector Control - Target=" + str(target) + ", Final=" + str(final))
plt.xlabel("X Coord (cm)")
plt.ylabel("Y Coord (cm)")
plt.xlim((x_max * -1, x_max))
plt.ylim((-1, -1 + y_max))
plt.legend()
plt.savefig("controltarg_" + str(target[0]) + "_" + str(target[1]) + ".png")
plt.show()
```
