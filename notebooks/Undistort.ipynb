```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
plt.rcParams['figure.dpi'] = 200
```

```python
img = cv2.imread('/media/user1/Data 2000/soft_robotics_experiments/module_2_single_actuator_right/m2_right_actuator_simple3/0.jpg')
```

```python
def showImagePlt(title, img_param):
    rgb_img = cv2.cvtColor(img_param, cv2.COLOR_BGR2RGB)
    plt.imshow(rgb_img)
    plt.title(title)
    plt.show()
```

```python
showImagePlt("raw", img)
```

```python
mtx = np.array([[5.099854774938199284e+02,0.000000000000000000e+00,3.229310679103935513e+02],
                [0.000000000000000000e+00,5.111052180009754125e+02,2.366653358694009057e+02],
                [0.000000000000000000e+00,0.000000000000000000e+00,1.000000000000000000e+00]])
dist = np.array([2.041200575421323049e-01,-3.984469142078214676e-01,-2.187974800224562911e-03,2.996656098133694632e-04,5.788197147242510332e-02])
```

```python
h, w = img.shape[:2]
print(h)
print(w)
```

```python
newcameramtx, roi = cv2.getOptimalNewCameraMatrix(mtx, dist, (w,h), 1, (w,h))
```

```python
newcameramtx
```

```python
undistorted_img = cv2.undistort(img, mtx, dist, None, newcameramtx)
```

```python
showImagePlt("raw", undistorted_img)
```

```python
und = cv2.undistort(img, mtx, dist, newCameraMatrix=mtx)
```

```python
showImagePlt("und", und)
```
