```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
plt.rcParams['figure.dpi'] = 100
experiment_count = 3
```

```python
exp_dir = "/media/user1/Data 2000/soft_robotics_experiments/module_2_single_actuator_right/"
exp_1_dir = exp_dir + "m2_right_actuator_simple"
exp_2_dir = exp_dir + "m2_right_actuator_simple2"
exp_3_dir = exp_dir + "m2_right_actuator_simple3"

exp_1_data = exp_1_dir + ".csv"
exp_1_markers = exp_1_dir + "_markers.csv"
exp_1_poly = exp_1_dir + "_poly.csv"

exp_2_data = exp_2_dir + ".csv"
exp_2_markers = exp_2_dir + "_markers.csv"
exp_2_poly = exp_2_dir + "_poly.csv"

exp_3_data = exp_3_dir + ".csv"
exp_3_markers = exp_3_dir + "_markers.csv"
exp_3_poly = exp_3_dir + "_poly.csv"

exp_df = []
markers_df = []


exp_df.append(pd.read_csv(exp_1_data))
exp_df.append(pd.read_csv(exp_2_data))
exp_df.append(pd.read_csv(exp_3_data))

markers_df.append(pd.read_csv(exp_1_markers))
markers_df.append(pd.read_csv(exp_2_markers))
markers_df.append(pd.read_csv(exp_3_markers))
```

```python
sample_counts = []
for i in range(experiment_count):
    count = min(exp_df[i].shape[0], markers_df[i].shape[0])
    print(count)
    sample_counts.append(count)
    exp_df[i] = exp_df[i][:count]
    markers_df[i] = markers_df[i][:count]   
```

# Pressure vs Time in each experiment

```python
for i in range(experiment_count):
    time = exp_df[i]["TIME"]
    pressure = exp_df[i]["M2-PR"]
    plt.plot(time, pressure, label="Run #" + str(i))
    
plt.title("Pressure vs Time in the Right Actuator")
plt.ylabel("Pressure (kPa)")
plt.xlabel("Time (sec)")
plt.legend()
```

# X & Y Coordinates vs Time in each experiment

```python
for i in range(experiment_count):
    time = exp_df[i]["TIME"]
    x = markers_df[i]["M10X"]
    plt.plot(time, x, label="Run #" + str(i))
    
plt.title("X-value vs Time in the End Marker")
plt.ylabel("Pixels")
plt.xlabel("Time (sec)")
plt.legend()
```

```python
for i in range(experiment_count):
    time = exp_df[i]["TIME"]
    y = markers_df[i]["M10Y"]
    plt.plot(time, y, label="Run #" + str(i))
    
plt.title("Y-value vs Time in the End Marker")
plt.ylabel("Pixels")
plt.xlabel("Time (sec)")
plt.legend()
```

# Pressure Difference vs changes in X&Y-Coordinate

```python
for i in range(experiment_count):
    pressure_difference = exp_df[i]["M2-PR"] - exp_df[i]["M2-PL"]
    
    x_start = markers_df[i]["M5X"][0]
    x_delta = (markers_df[i]["M10X"] - x_start)*-1
    
    plt.plot(pressure_difference, x_delta, "x", label="Run #" + str(i))
    
plt.title("Change in X vs Pressure difference")
plt.ylabel("Pixels")
plt.xlabel("Pressure (kPa)")
plt.legend()
```

```python
for i in range(experiment_count):
    pressure_difference = exp_df[i]["M2-PR"] - exp_df[i]["M2-PL"]
    
    y_start = markers_df[i]["M10Y"][0]
    y_delta = (markers_df[i]["M10Y"] - y_start)
    
    plt.plot(pressure_difference, y_delta, "x", label="Run #" + str(i))
    
plt.title("Change in Y vs Pressure difference")
plt.ylabel("Pixels")
plt.xlabel("Pressure (kPa)")
plt.legend()
```
